```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the raw dataset
raw_data_path = "../data/raw_data.csv"
raw_data = pd.read_csv(raw_data_path)

# Display the first few rows of the dataset
raw_data.head()
```

```python
summary_stats = raw_data.describe()
```

```python
raw_data.info()
```

```python
# Explore unique value in categorical column 
categorical_columns = raw_data.select_dtypes(include="object").columns
for column in categorical_columns:
    print(f"Unique values in {column}: {raw_data[column].unique()}")

```

```python
# Visualize data distributions
for column in raw_data.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(raw_data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Visualize correlations between numerical features
numeric_columns = raw_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = raw_data[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
